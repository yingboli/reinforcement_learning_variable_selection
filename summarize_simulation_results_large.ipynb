```python
import pandas as pd
import numpy as np
```

## Combine results 

```python
## n = 10000
results = pd.read_csv("results_toeplitz/toeplitz_design_results.csv")
## Filter to only keep n=10000, and run_id <= 4
results = results.loc[(results['n'] == 10000) & (results['run_id'] <= 4)]
results2 = pd.read_csv("results_toeplitz_large/toeplitz_design_results.csv")
## Filter to only keep n=10000, and run_id <= 3
results2 = results2.loc[(results2['n'] == 10000) & (results2['run_id'] <= 3)]

results2['run_id'] = results2['run_id'] + 10
results = pd.concat([results, results2])
n_repeat = 7 # n=10000
```

```python
## n = 1000

results = pd.read_csv("results_toeplitz/toeplitz_design_results.csv")
## Filter to only keep n=10000, and run_id <= 4
results = results.loc[(results['n'] == 1000) & (results['run_id'] <= 4)]
results2 = pd.read_csv("results_toeplitz_small/toeplitz_design_results.csv")

results2['run_id'] = results2['run_id'] + 10
results = pd.concat([results, results2])
```

```python
n_repeat = len(results['run_id'].unique())
```

```python
n_repeat
```

```python
group_cols = ["n", "p_total", "p_true", "snr", "rho", "reward_type"]
agg_spec = {}
sequential_gammas = [0.3]
_methods = ["bandit"] + [f"sequential_g{g}" for g in sequential_gammas] + ["lasso", "adaptive_lasso", "forward", "backward", "mcmc"]
for method in _methods:
    for m in ["n_selected", "test_r2", "test_mse", "f1", "runtime_sec"]:
        col = f"{method}_{m}"
        agg_spec[f"{col}_mean"] = (col, "mean")
        agg_spec[f"{col}_std"] = (col, "std")
agg_spec["all_features_test_r2_mean"] = ("all_features_test_r2", "mean")
agg_spec["all_features_test_r2_std"] = ("all_features_test_r2", "std")
results = results.groupby(group_cols, as_index=False).agg(**agg_spec)
```

## Start Analyzing

```python
n = 1000
p_total = 50
p_true = 20
snr = 0.5
rho = 0.9

current_results = results.loc[(results['n'] == n) & (results['p_total'] == p_total) & (results['p_true'] == p_true) & (results['snr'] == snr) & (results['rho'] == rho) ]
```

```python
metric = 'f1'
models = ['bandit', 'sequential_g0.3', 'lasso', 'adaptive_lasso', 'forward', 'backward', 'mcmc']

results_mean = current_results[['reward_type'] + [x + '_' + metric + '_mean' for x in models]].set_index('reward_type').loc[['cv_rmse', 'aic', 'bic', 'bayes_factor']]
results_mean.columns = models
results_mean.index = ['CV-RMSE', 'AIC', 'BIC', 'Bayes Factor']
results_sd = current_results[['reward_type'] + [x + '_' + metric + '_std' for x in models]].set_index('reward_type').loc[['cv_rmse', 'aic', 'bic', 'bayes_factor']] / np.sqrt(n_repeat)
results_sd.columns = models
results_sd.index = ['CV-RMSE', 'AIC', 'BIC', 'Bayes Factor']
round(results_mean, 4)
# round(results_sd, 2)
print(results_mean.style.format(precision=2).to_latex())
```
